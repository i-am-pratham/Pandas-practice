# #make a new column for episode count

```python
# first we have to make new column for "episode count". but in this csv there are only 3 column which are Rank,Title and Score
# so we have to extract episode count from Title .
```

```python
# We can see episode count are there in Title , now we have to extract it.
```

```python
def episode_count(txt):
    check=False
    data=''
    for i in txt:
       
        if i==')':
            check=False
            return data
            break

        if check==True:
            data=data+i
                   
        if i== '(':
            check= True
            
        # using this function we extract the episodes.
        
```

```python
df['Episodes']=df["Title"].apply(episode_count) #apply the function
```

```python
df["Episodes"]=df["Episodes"].str.replace(' eps','')  #remove "eps" and only keep numbers for further calculation
```

# #make a new column for time stamp

```python
df.loc[1]["Title"]
```

```python
def extract_time(txt):
    data=''
    for i in range(len(txt)):
        if txt[i] == ')':
            for j in range(i+1,i+20):
                data=data+txt[j]
            return data
```

```python
df['Total Time']=df['Title'].apply(extract_time)
            
```
